```python
#Calculator useing single inheritance

class calculator:
  def __init__(self,num1, num2):
    self.num1 = num1
    self.num2 = num2

class calcFunction(calculator):
  def add(self):
    return self.num1 + self.num2

  def sub(self):
    return self.num1 - self.num2

  def mul(self):
    return self.num1 * self.num2

  def div(self):
    return self.num1 / self.num2

while True:
  print("\nChoose Operations:")
  print("1. Addition")
  print("2. Subtraction")
  print("3. Multiplication")
  print("4. Division")
  print("5. Exit")

  choice = int(input("Enter your choice: "))

  if choice == 5:
    print("Exiting...")
    break

  if choice in [1, 2, 3, 4]:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    cal = calcFunction(num1,num2)

    if choice == 1:
      print(f"Additon of {num1, num2} given number is: ", cal.add())
    elif choice == 2:
      print(f"Subtraction of {num1, num2} given number is: ",  cal.sub())
    elif choice == 3:
      print(f"Multiplication of {num1, num2} given number is: ", cal.mul())
    elif choice == 4:
      print(f"Division of {num1, num2} given number is: ", cal.div())
  else:
    print("Invalid choice")
```

```python
class NumberContainer:
    def __init__(self, values):
        self.values = values

class SumCalculator(NumberContainer):
    def __init__(self, values):
        NumberContainer.__init__(self, values)

    def calculate_total(self):
        total = 0
        for number in self.values:
            total += number
        return total

values = list(range(1, 101))
calculator = SumCalculator(values)
total_sum = calculator.calculate_total()

print(f"The sum of the numbers is: {total_sum}")
```
