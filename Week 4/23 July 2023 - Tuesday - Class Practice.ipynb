```python
#Q1. Create a program of finding parimeter of a triangle and area of any three shapes using class object
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def calculate_perimeter(self):
        return self.a + self.b + self.c

    def calculate_area(self, height):
        base = self.a
        return 0.5 * base * height

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        return 3.14159 * self.radius * self.radius

def main():
    print("Choose an option:")
    print("1. Calculate perimeter of a triangle")
    print("2. Calculate area of a triangle")
    print("3. Calculate area of a rectangle")
    print("4. Calculate area of a circle")
    print()

    choice = input("Enter your choice (1/2/3/4): ")
    print()

    if choice == "1":
        a = float(input("Enter the first side of the triangle: "))
        b = float(input("Enter the second side of the triangle: "))
        c = float(input("Enter the third side of the triangle: "))
        triangle = Triangle(a, b, c)
        perimeter = triangle.calculate_perimeter()
        print(f"The perimeter of the triangle is: {perimeter}")

    elif choice == "2":
        a = float(input("Enter the base of the triangle: "))
        b = float(input("Enter the second side of the triangle: "))
        c = float(input("Enter the third side of the triangle: "))
        height = float(input("Enter the height of the triangle: "))
        triangle = Triangle(a, b, c)
        area = triangle.calculate_area(height)
        print(f"The area of the triangle is: {area}")

    elif choice == "3":
        length = float(input("Enter the length of the rectangle: "))
        width = float(input("Enter the width of the rectangle: "))
        rectangle = Rectangle(length, width)
        area = rectangle.calculate_area()
        print(f"The area of the rectangle is: {area}")

    elif choice == "4":
        radius = float(input("Enter the radius of the circle: "))
        circle = Circle(radius)
        area = circle.calculate_area()
        print(f"The area of the circle is: {area}")

    else:
        print("Invalid choice, please choose 1, 2, 3, or 4.")

main()
```

```python
#Q2. Print the sum of 1000 numbers using class object concepts

class sumCalculator:
  def __init__(self, n):
    self.n = n

  def calculateSum(self):
    total = 0
    for i in range(1, self.n + 1):
      total += i
    return total

calculate = sumCalculator(1000)
result = calculate.calculateSum()
print("Sum of 1000 numbers is: ", result)
```

```python
#Q3. Create a factorial program using class object
class factorial:
  def __init__(self, n):
    self.n = n

  def calculateFactorial(self):
    if self.n == 0 or self.n ==1:
      return 1
    else:
      factorial = 1
      for i in range(2, self.n + 1):
        factorial *= i
      return factorial

n = int(input("Enter a number: "))

calculate = factorial(n)
result = calculate.calculateFactorial()
print("Factorial of", n, "is: ", result)
```

```python
#Q4. create a fibonacci series using class object
class Fibonacci:
    def __init__(self, n):
        self.n = n

    def generate_series(self):
        series = [0] * self.n
        a, b = 0, 1
        for i in range(self.n):
            series[i] = a
            a, b = b, a + b
        return series

n = int(input("Enter the number of terms in the Fibonacci series: "))

fibonacci = Fibonacci(n)

fibonacci_series = fibonacci.generate_series()
print(f"The first {n} terms of the Fibonacci series are: {fibonacci_series}")
```
