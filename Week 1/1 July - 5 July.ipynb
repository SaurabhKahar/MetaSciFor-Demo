Create a Atm program using while loop major one

```python
#ATM Program

balance = 10000         #Initial amount stored
pin = "12345"        #Password

while True:
  print("Choose Options:")
  print("1. Check Balance")
  print("2. Deposit")
  print("3. Withdraw")
  print("4. Exit")

  choice = input("Enter your choice: ")       #User input for what he/she wants to execute

  if choice == "1":                           #for checking balance
    user_pin = input("Enter your pin: ")
    if user_pin == pin:
      print("Your balance is: ", balance)
    else:
      print("Incorrect pin. Please try again.")

  elif choice == "2":                         #for depositing
    user_pin = input("Enter your pin: ")
    if user_pin == pin:
      amount = int(input("Enter the amount to deposit: "))
      balance = balance + amount
      print("Deposit successful. \nNew balance: ", balance)
    else:
      print("Incorrect pin. Please try again.")

  elif choice == "3":                         #for withdrawing
    user_pin = input("Enter your pin: ")
    if user_pin == pin:
      amount = int(input("Enter the amount to withdraw: "))
      if amount <= balance:
        balance = balance - amount
        print("Withdrawal successful. \nNew balance: ", balance)
      else:
        print("Invalid amount or Insufficient amount.")
    else:
      print("Incorrect pin. Please try again.")

  elif choice == "4":                         #for exiting
    print("Thank you for visiting ATM.")
    break

  else:
    print("Invalid choice. Please try again.")    # if user selected wrong choice
```
